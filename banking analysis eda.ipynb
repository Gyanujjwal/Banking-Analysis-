```python
for col in categorical_cols:
    if col == "Occupation":
        continue
    
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()
```

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

# Univariate analysis and visualization
plt.figure(figsize=(15,10))

for i, col in enumerate(numerical_cols):
    plt.subplot(4, 3, i+1)
    sns.histplot(df[col], kde=True)
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
## Heatmaps

numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix,annot=True,cmap='crest',fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

## EDA INSIGHTS OF THE ANALYSIS 
1. Checking Account vs Deposit
Checking accounts are mostly used by salaried individuals or business people.

Customers with a checking account usually have moderate to high deposits, as they regularly use the account for transactions.

However, having a checking account doesn’t always mean high deposits — it's often just a basic necessity.

2. Savings Account vs Bank Deposits
Customers with savings accounts tend to maintain consistent but moderate deposits.

These accounts are not used for frequent transactions, so balances remain stable.

Often linked with salary or long-term saving goals, especially in middle-income groups.

Trend: Savings account holders show gradual growth in deposits over time.

3. Foreign Currency Account vs Bank Deposits
Customers with foreign currency accounts tend to hold high deposit values, often in USD, EUR, etc.

These accounts are commonly used by NRIs, exporters/importers, or HNI (High Net-worth Individuals).

The exchange rate advantage often encourages higher deposits.

Trend: Strong positive correlation between foreign currency account ownership and total deposit amount.
